# Introduction:- 

Diamond Sales Data Analysis
In today’s competitive luxury goods market, diamonds are among the most sought-after products, with their sales driven by various factors such as customer demographics, product preferences, and purchasing behavior. To better understand these factors, this project aims to analyze and model the sales data of diamond products. The dataset used in this project includes detailed information about diamond sales transactions, which provides an opportunity to extract insights into customer buying patterns, popular product categories, and geographical distribution of sales.

This dataset consists of several key features that capture a variety of dimensions of the diamond sales process, such as:

Transaction Information: This includes Transaction_ID, Date_of_Sale, and Payment_Method, providing insights into the time period and the mode of payment preferred by customers.
Customer Demographics: Attributes such as Customer_Age, Customer_Gender, and geographic information like State and City allow for an analysis of the customer base and their purchasing behavior across different regions.
Product Specifications: Data about the Product_Category (e.g., Ring, Necklace, Bracelet), Carat, Diamond_Clarity, and Diamond_Color provides insights into the types of diamond products sold, allowing us to understand product preferences.
Sales Metrics: The Price column records the amount spent on each transaction, and Discount_Applied represents the discount offered on the product, both of which will help in revenue and pricing analysis.

# Dataset Features:-

The dataset contains the following columns:

Transaction_ID: A unique identifier for each transaction.
Date_of_Sale: The date on which the transaction occurred.
Customer_Age: Age of the customer at the time of the transaction.
Customer_Gender: Gender of the customer (Male or Female).
State: The state in which the customer resides.
City: The city in which the customer resides.
Product_Category: The category of the diamond product (e.g., Ring, Bracelet, Necklace).
Carat: The weight of the diamond in carats.
Diamond_Clarity: The clarity grade of the diamond (e.g., VVS1, SI1).
Diamond_Color: The color grade of the diamond (e.g., D, F, G).
Price: The price of the diamond product.
Payment_Method: The payment method used in the transaction (e.g., Online, Credit Card, Cash).
Discount_Applied: The discount percentage applied to the transaction.

# 6). Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# Selecting features and target variable
features = df_diamond[['Carat', 'Customer_Age', 'Diamond_Clarity', 'Customer_Gender', 'State', 'Product_Category', 'Sale_Month']]
target = df_diamond['Price']

# Appling one-hot encoding to all categorical columns
features_encoded = pd.get_dummies(features, columns=['Diamond_Clarity', 'Customer_Gender', 'State', 'Product_Category', 'Sale_Month'], drop_first=True)

# Spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=0.2, random_state=42)

# Models to evaluate
models = {
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42),
    'LinearRegression': LinearRegression(),
    'DecisionTree': DecisionTreeRegressor(random_state=42),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=100, random_state=42)
}

# Dictionary to store the MSE of each model
model_performance = {}

# Train and evaluate each model
for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    predictions = model.predict(X_test)  # Make predictions
    mse = mean_squared_error(y_test, predictions)  # Evaluate the model
    model_performance[model_name] = mse  # Store the MSE
    print(f'{model_name} - Mean Squared Error: {mse:.2f}')

# Comparing the results
print("\nModel Performance Comparison:")
for model_name, mse in model_performance.items():
    print(f"{model_name}: {mse:.2f}")
```

Based on the results of the four models, here's a conclusion:-

Linear Regression performed the best, with the lowest Mean Squared Error (MSE) of 9.39 billion. This suggests that the relationship between the features and the target variable (price) can be well-approximated by a linear model.

Gradient Boosting Regressor also performed quite well, with an MSE of 9.49 billion, showing it can handle more complex relationships slightly better than RandomForest.

Random Forest Regressor had a moderately higher MSE of 10.26 billion, indicating that while it captures some non-linearities, it may not generalize as well as Gradient Boosting or Linear Regression in this case.

Decision Tree Regressor had the highest MSE at 19.51 billion, indicating that it overfits the data and performs poorly compared to the other models.

In conclusion, Linear Regression and Gradient Boosting are the most suitable models for predicting diamond prices based on the given features, with Linear Regression slightly outperforming other models. However, Gradient Boosting offers a good balance for capturing complex patterns while avoiding overfitting.
